```python
def age_range(x):
    x = x[1:]
    return x
tbmelt['age'] = tbmelt['agesex'].apply(lambda a: age_range(a))
```

```python
def age_range_hyphen(x):
    if x == '65':
        x = x + '-'
    elif len(x) == 2 or len(x) == 3:
        x = x[0] + '-' + x[1:]
    elif len(x) == 4:
        x = x[:2] + '-' + x[2:]
    return x

tbmelt['age'] = tbmelt['age'].apply(lambda a: age_range_hyphen(a))
```
